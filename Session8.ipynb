```python
import matplotlib.pyplot as plt
import numpy as np

#Array from [0,1] in 100 steps
arrX = np.linspace(0,1,100)

#Creating variables to hold the values of sin(x) and cos(x)
y = np.sin(arrX)
z = np.cos(arrX)
```

```python
#Call subplots to generate a multipanel figure. This means 1 row, 2 columns of figures
f, multiPanel = plt.subplots(1,2)

#First Panel
multiPanel[0].plot(arrX,y)
multiPanel[0].set_xlabel("x")
multiPanel[0].set_ylabel("sin(x)")
multiPanel[0].set_title(r"$\sin(x)$")

#Second Panel
multiPanel[1].plot(arrX,z)
multiPanel[1].set_xlabel("x")
multiPanel[1].set_ylabel("cos(x)")
multiPanel[1].set_title(r"$\cos(x)$")

#Adjust Subplots
f.subplots_adjust(wspace = 0.5)
plt.savefig("session8.pdf", facecolor="white", dpi=600, bbox_inches="tight")
```
